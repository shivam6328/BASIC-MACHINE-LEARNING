```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
x=pd.read_csv('Downloads/iris.csv')
x.columns=['sepal length','sepal width','petal length','petal width','Class']
x
```

```python
encode=LabelEncoder()
x.Class=encode.fit_transform(x.Class)
x
```

```python
train,test=train_test_split(x,test_size=0.33)
```

```python
train_x=train.drop(columns=['Class'],axis=1)
train_y=train['Class']
test_x=test.drop(columns=['Class'],axis=1)
test_y=test['Class']
```

```python
lnr=LogisticRegression()
ans=lnr.fit(train_x,train_y)
```

```python
gnb=GaussianNB()
ans1=gnb.fit(train_x,train_y)
predict=gnb.predict(test_x)
print accuracy_score(test_y,predict)
```
